**Funcion input()**
con la funcion input podemos interactuar con nuestro programa, el mismo se detiene momentáneamente y espera a que el usuario ingrese algún texto a través del teclado.
Una vez que el usuario ha proporcionado la entrada y presionado la tecla "Enter", input() recoge esa entrada y la almacena en una variable que podemos utilizar más adelante.
*Dato importante: Siempre el valor es de tipo str*

```python
# ejemplo de input
nombre = input("Ingrese su nombre: ")
print(nombre)
```

```python
# ejemplo de input con int, para que vean que ingresa todo como str
numero = input("Ingrese un numero: ")
print(type(numero))
```
